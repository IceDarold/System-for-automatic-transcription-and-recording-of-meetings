```python
import os
import json
import torch
from pyannote.audio import Pipeline
from pydub import AudioSegment
from datetime import timedelta
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

class AudioDiarizer:
    def __init__(self, hf_token: str = None):
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        logger.info(f"Используемое устройство: {self.device}")

        self.hf_token = hf_token or os.getenv("HF_TOKEN")
        if not self.hf_token:
            raise ValueError()
        try:
            self.pipeline = Pipeline.from_pretrained(
                "pyannote/speaker-diarization-3.1",
                use_auth_token=self.hf_token
            ).to(self.device)
            logger.info("Модель диаризации успешно загружена")
        except Exception as e:
            logger.error(f"Ошибка загрузки модели: {str(e)}")
            raise

    def process_audio(
        self,
        input_path: str,
        output_dir: str = "output",
        min_segment_duration: float = 0.5
    ) -> dict:
        try:
            # Проверка входного файла
            if not os.path.isfile(input_path):
                raise FileNotFoundError(f"Файл не найден: {input_path}")

            # Создание папки для результатов
            os.makedirs(output_dir, exist_ok=True)
            logger.info(f"Создана папка для результатов: {output_dir}")

            # Загрузка аудио
            audio = AudioSegment.from_file(input_path)
            duration = len(audio) / 1000  # Длительность в секундах
            logger.info(f"Загружен аудиофайл: {input_path} (длительность: {duration:.2f} сек)")

            # Диаризация
            logger.info("Начало диаризации...")
            diarization = self.pipeline(input_path)

            # Обработка результатов
            results = {
                "audio_file": os.path.basename(input_path),
                "duration_seconds": duration,
                "sample_rate": audio.frame_rate,
                "speakers": {},
                "segments": []
            }

            for segment, _, speaker in diarization.itertracks(yield_label=True):
                seg_duration = segment.end - segment.start
                if seg_duration < min_segment_duration:
                    continue

                # Форматирование времени
                start_str = str(timedelta(seconds=segment.start)).split('.')[0]
                end_str = str(timedelta(seconds=segment.end)).split('.')[0]

                # Данные сегмента
                segment_data = {
                    "speaker": speaker,
                    "start": round(segment.start, 2),
                    "end": round(segment.end, 2),
                    "duration": round(seg_duration, 2),
                    "start_time": start_str,
                    "end_time": end_str
                }
                results["segments"].append(segment_data)

                # Статистика по спикерам
                if speaker not in results["speakers"]:
                    results["speakers"][speaker] = {
                        "total_segments": 0,
                        "total_duration": 0.0
                    }
                results["speakers"][speaker]["total_segments"] += 1
                results["speakers"][speaker]["total_duration"] += seg_duration

            # Сохранение результатов
            self._save_results(results, output_dir, audio)
            logger.info("Обработка завершена успешно")
            return results

        except Exception as e:
            logger.error(f"Ошибка обработки аудио: {str(e)}", exc_info=True)
            raise

    def _save_results(self, results: dict, output_dir: str, audio: AudioSegment):
        """Сохранение результатов в файлы"""
        # 1. JSON с полной информацией
        json_path = os.path.join(output_dir, "segments.json")
        with open(json_path, "w", encoding='utf-8') as f:
            json.dump(results, f, indent=2, ensure_ascii=False)

        # 2. Текстовый файл с метками
        txt_path = os.path.join(output_dir, "segments.txt")
        with open(txt_path, "w", encoding='utf-8') as f:
            f.write(f"Файл: {results['audio_file']}\n")
            f.write(f"Общая длительность: {results['duration_seconds']:.2f} сек\n")
            f.write(f"Частота дискретизации: {results['sample_rate']} Гц\n")
            f.write(f"Количество спикеров: {len(results['speakers'])}\n\n")

            for speaker, stats in results["speakers"].items():
                f.write(f"Спикер {speaker}:\n")
                f.write(f"  Всего сегментов: {stats['total_segments']}\n")
                f.write(f"  Общее время: {stats['total_duration']:.2f} сек\n\n")

            f.write("Детализация сегментов:\n")
            f.write("-" * 70 + "\n")

            for seg in results["segments"]:
                f.write(
                    f"Спикер: {seg['speaker']}\n"
                    f"Время: {seg['start_time']} - {seg['end_time']}\n"
                    f"Длительность: {seg['duration']:.2f} сек\n"
                    f"Точные метки: {seg['start']:.2f}-{seg['end']:.2f} сек\n"
                    f"-" * 70 + "\n"
                )

        # 3. Сохранение аудио сегментов (опционально)
        segments_dir = os.path.join(output_dir, "audio_segments")
        os.makedirs(segments_dir, exist_ok=True)

        for seg in results["segments"]:
            start_ms = int(seg["start"] * 1000)
            end_ms = int(seg["end"] * 1000)
            segment_audio = audio[start_ms:end_ms]
            filename = f"{seg['speaker']}_{seg['start']:.1f}-{seg['end']:.1f}s.wav"
            segment_audio.export(os.path.join(segments_dir, filename), format="wav")


if __name__ == "__main__":
    try:
        diarizer = AudioDiarizer()
        results = diarizer.process_audio(
            input_path="НАЗВАНИЕ ФАЙЛА",
            output_dir="diarization_results",
            min_segment_duration=1.0
        )

        # Вывод краткой статистики
        print("\nРезультаты диаризации:")
        print(f"Файл: {results['audio_file']}")
        print(f"Общая длительность: {results['duration_seconds']:.2f} сек")
        print(f"Найдено спикеров: {len(results['speakers'])}")

        for speaker, stats in results["speakers"].items():
            print(f"\nСпикер {speaker}:")
            print(f"  Сегментов: {stats['total_segments']}")
            print(f"  Общее время: {stats['total_duration']:.2f} сек")

        print("\nПодробные результаты сохранены в папке 'diarization_results'")

    except Exception as e:
        logger.error(f"Критическая ошибка: {str(e)}")
        exit(1)
```
