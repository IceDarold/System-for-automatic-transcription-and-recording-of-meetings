```python
import whisper
from moviepy.editor import VideoFileClip
import os
import noisereduce as nr
import soundfile as sf
import librosa

audio, sr = librosa.load("lukash.mp3", sr=16000)
noisy_part = audio[5000:15000]
reduced_noise = nr.reduce_noise(y=audio, y_noise=noisy_part, sr=sr)
processed_audio = audio
sf.write("output.wav", processed_audio, sr)

def transcribe_audio(audio_path, model_size="small", language="ru"):
    model = whisper.load_model(model_size)
    result = model.transcribe(audio_path, language=language)
    return result["text"]

if __name__ == "__main__":
    try:
        text = transcribe_audio("output.wav")
        print("Расшифрованный текст:")
        print(text)
        with open("transcription.txt", "w", encoding="utf-8") as f:
            f.write(text)
        print("\nТекст сохранён в 'transcription.txt'")

    except Exception as e:
        print(f"Ошибка: {e}")
```
