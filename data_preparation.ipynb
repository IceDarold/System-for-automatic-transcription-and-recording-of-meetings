```
import noisereduce as nr
import soundfile as sf
import librosa

# Load audio file with a target sample rate of 16 kHz
audio, sr = librosa.load("test.ogg", sr=16000)

# Select a segment that contains only noise
noisy_part = audio[5000:15000]

# Perform noise reduction
reduced_noise = nr.reduce_noise(y=audio, y_noise=noisy_part, sr=sr)

# Save the processed audio to output.wav
sf.write("output.wav", reduced_noise, sr)
```
